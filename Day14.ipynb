# Q1. Problem Statement: Measures of Central Tendency- I  
Write a Python program to declare a 1-D array. With the help of the statistics 
library, calculate the Mean, Median, and Mode. 
Note: You can write any data of your choice inside the array 

```python
#import the relevant libraries
import numpy as np
import scipy as stats
from statistics import mode

#A program using a loop to get a list of values as our data
def Get_Central_Tendencies():
    y = int(input(f"Enter the number of values: "))
    data = []
    for x in range(y):
        value = float(input(f"Enter value {x+1}: "))
        data.append(value)
    print("The Array is ",data)
    
    mean_value = np.mean(data)
    print("Mean of the Array is: ", mean_value)
    
    median_value = np.median(data)
    print("Median of the Array is: ", median_value)
    
    mode_value = mode(data)
    print("Mode of the Array is: ", mode_value)
    return
```

```python
Get_Central_Tendencies()
```

Load the “Books.csv” dataset into a DataFrame and perform the following tasks: 
1. Perform initial analysis on the dataset using the info() and the 
describe() functions 
2. Group the data based on the "Author" column and calculate the average user 
ratings and average price, and write them in new columns 
3. Find out which of the authors have written the maximum number of books 

```python
import pandas as pd

df2 = pd.read_csv("Day14-Dataset/Books.csv")

df2.head()
```

```python
#performing initial analysis
df2.info()
```

```python
df2.describe()
```

```python
#Group the data based on the "Author" column and calculate the average user ratings and average price, and write them in new columns
grouped_data = df2.groupby('Author')
summed_values = grouped_data[['User Rating','Price']].mean()
summed_values
```

```python
#Find out which of the authors have written the maximum number of books
grouped_data = df2.groupby('Author').agg({'Name': 'count'})

#Rename the columns
grouped_data = grouped_data.rename(columns={'Author': 'Author', 'Name':'Book Count'})

#Sort Values in descending order
grouped_data = grouped_data.sort_values(by='Book Count', ascending=False)

#Finding the author with max number of books
max_books_count = grouped_data['Book Count'].max()
authors = grouped_data[grouped_data['Book Count'] == max_books_count]
authors
```
