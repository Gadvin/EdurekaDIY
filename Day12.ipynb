# Q1. Problem Statement: Sales Data Analysis- I 
Load the “sample_superstore.csv” dataset provided on LMS into a DataFrame, 
using the Matplotlib library, plot the following insights from the dataset: 
1. Compare the total sales for each category of products 
2. Number of deals in each state of America 
Note: You can use any plot of your choice to show the insights 

```python
import pandas as pd

df = pd.read_csv("Day12-I-Dataset/SampleSuperstore.csv")
```

```python
df.head(100)
```

```python
import matplotlib.pyplot as plt
```

```python
#ploting using the bar graph from matplotlib to Compare the total sales for each category of products
plt.bar(df.Category, df.Sales)

#adding the title and labels
plt.title('Total Sales for Each Category')
plt.xlabel('Category')
plt.ylabel('Sales')

#Displaying the graph
plt.show()
```

```python
#ploting using the bar graph from matplotlib Number of deals in each state of America

#Find the value counts of the each State
state_deals = df['State'].value_counts()

#ploting the bar graph using the above information in a defined figure size
plt.figure(figsize=(12,10))
plt.bar(state_deals.index, state_deals, color='lightgreen')

#adding the size,title and labels
plt.title('Total Number of Deals in Each American State')
plt.xlabel('State')
plt.xticks(rotation=90)
plt.ylabel('Deals')

#Displaying the graph
plt.show()
```

# Q2. Problem Statement: Data Grouping - II 
Load the “sample_superstore.csv” dataset provided on LMS into a DataFrame; 
using the Seaborn library, plot the following insights from the dataset: 
1. Print a correlation matrix between Postal code, Quantity, Sales, Discount, and Profit 
2. Plot the correlation matrix using a heatmap 

Note: You can use any plot of your choice to show the insights

```python
#importing the seaborn library and load the data to a dataframe
import seaborn as sns

df1 = pd.read_csv("Day12-I-Dataset/SampleSuperstore.csv")

#get the first 5 rows of the data
df1.head()
```

```python
#get the correlation matrix between Postal Code, Quantity, Sales, Discount and Profit
#subset to get the columns of interest
df2 = df1[['Postal Code', 'Quantity', 'Sales', 'Discount','Profit']]

#gettig the correlatio matrix
corr_matrix = df2.corr()
print(corr_matrix)

```

```python
#plot the heatmap using the seaborn library
plt.figure(figsize=(10, 8)) #Setting the figure sie
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1, vmin=-1, vmax=1)

#adding the title
plt.title('Correlation between Postal code, Quantity, Sales, Discount, and Profit')

#show the graph
plt.show()
```
