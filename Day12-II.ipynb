# Q1. Problem Statement: Data Visualization using Matplotlib 
Load the “Netflix_titles.csv” data into a DataFrame and plot the following 
insights using the Matplotlib library: 
1. How much of the Netflix content are TV shows or Movies? 
2. In which country is Netflix the most popular?

```python
import pandas as pd

df = pd.read_csv("Day12-II-Dataset/netflix_titles.csv")

df.head()
```

```python
#importing matplotlib
import matplotlib.pyplot as plt
```

```python
#plot a pie char to show the type of content

#getting the value counts of the types
types_values = df['type'].value_counts()
print(types_values)

#plotting the pie chart to show the ratios
#plt.pie(types_values, explode=(0.1,0), labels=[ 'Movies','TV Shows'])
```

```python
#plot a pie char to show the type of content

#getting the value counts of the types
country_values = df['country'].value_counts()
top_countries = country_values.head(14)

#plotting the pie chart to show the ratios
plt.pie(top_countries, labels = top_countries.index)

#adding the title
plt.title('NetFlix Viewers Accross the Globe')
```
