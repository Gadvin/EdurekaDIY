# Q1. Problem Statement: Data Filtering using Pandas

Write a Python program that reads the player.csv (provided on LMS) file into a 
DataFrame, then shows only those players who played more than 150 innings 
and scored at least two centuries and 35 fifties.

```python
import pandas as pd

df = pd.read_csv("DIY10/player.csv")
```

```python
df.head(100)
```

```python
#filtered data based on the players who played more than 150 innings and scored at least two centuries and 35 fifties.
filtered_df = df[(df['innings']>150) & (df['100s']>2) & (df['50s']>35)]
```

```python
#Printing the output of the filtered data
print(filtered_df)
```

Use the above data set and do the below tasks.
1. Remove the data of players who didn't score a single century 
2. Replace the player name "V Kohli " with "Virat Kohli “ 
3. How many matches they weren't getting a chance for batting to fetch 
that and stored in a new column and named it as “not batting” 
4. Fetch data where they scored neither 100 nor 50 in matches and 
store it in the new column as renamed as “low score innings”

```python
df.head(100)
```

```python
df1 = df

print(df1.head())
```

```python
#Remove the data of players who didn't score a single century 
df1 = df[(df['100s']>0)]

df1
```

```python
#Replace the player name "V Kohli " with "Virat Kohli“

df1['name'] = df1['name'].replace('V Kohli', 'Virat Kohli')
df1
```

```python
#How many matches they weren't getting a chance for batting to fetch that and stored in a new column and named it as “not batting” 
#not battling = matches - innings
df1['not battling'] = df1['matches'] - df1['innings']
df1
```

```python
#Fetch data where they scored neither 100 nor 50 in matches and store it in the new column as renamed as “low score innings”
#low score inning = innings - (100s + 50s)
df1['Low score innings'] = df1['innings'] - (df1['100s'] + df['50s'])
df1
```
